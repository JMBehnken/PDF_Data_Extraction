# Extracting Data from a PDF

```python
import os
import pandas as pd
```

```python
path = '/Students_At_TU_Dortmund/'

files = sorted([file for file in os.listdir(path) if file.endswith('.txt')])
files
```

```python
n = 250

data = []
for file in files[-1:]:
    with open(path+'{}'.format(file), 'r') as f:
        i = 0
        while i<n:
            line = next(f).strip()
            if line.startswith('Tab. 1.1.1:'):
                    while i<n:
                        line = next(f).strip().replace('.', '')
                        if line.startswith('Insgesamt'):
                            break
                        if line:
                            data.append([element for element in line.split('  ') if element])
                        i += 1
            i += 1
```

```python
data_2 = []
columns = data[1]
columns.extend(data[2])
for i in range(1, len(data)):
    if len(data[i])==10:
        data_2.append(data[i])
    if len(data[i])==1 and len(data[i+1])==9 and len(data[i+2])==1:
        element = list([data[i][0]+' '+data[i+2][0]])
        element.extend(data[i+1])
        data_2.append(element)

df = pd.DataFrame(data_2[1:], columns=data_2[0])

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```
